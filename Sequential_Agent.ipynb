```python
from typing import TypedDict
from langgraph.graph import StateGraph
```

```python
class AgentState(TypedDict):
    name : str
    age : str
    final : str
```

```python
def first_node(state:AgentState) -> AgentState:
    """This is the first node of the sequence"""

    state["final"] = f"Hi {state["name"]}"
    return state

def second_node(state:AgentState) -> AgentState:
    """This is the second node of the sequence"""
    print("cascadss")
    state["final"] = state["final"] + f", You are {state["age"]} year old"
    return state
```

```python
graph = StateGraph(AgentState)

graph.add_node("First Step",first_node)
graph.add_node("Second Step",second_node)

graph.set_entry_point("First Step")
graph.add_edge("First Step","Second Step")
graph.set_finish_point("Second Step")

app = graph.compile()
```

```python
result = app.invoke({"name":"Apurv","age":"23"})

print(result)
```
