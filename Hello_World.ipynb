```python
from typing import Dict, TypedDict
from langgraph.graph import StateGraph
# Framework that helps you design and manage the flow of task in your application
```

```python
class AgentState(TypedDict): # Our Satate Schema
    message : str


def greeting_node(state : AgentState) -> AgentState:
    """ Simple node that add the greeting message to state"""
    state['message'] = "Hey" + state['message'] + ", How your day going"
    return state
```

```python
graph = StateGraph(AgentState)

graph.add_node("Greeter",greeting_node)

graph.set_entry_point("Greeter")

graph.set_finish_point("Greeter")

app = graph.compile()
```

```python
result = app.invoke({"message":" Apurv"})

print(result['message'])
```
