```python
from typing import TypedDict,List
from langgraph.graph import StateGraph,START,END
import random
```

```python
class AgentState(TypedDict):
    name : str
    data : List[int]
    counter : int

def greeting(state : AgentState) -> AgentState:
    """This function greets the user"""
    state['name'] = f"Hi there, {state['name']}"
    state["counter"] = 0
    return state

def randomNode(state: AgentState) -> AgentState:
    """Generate a random no from 0 to 10"""
    state['data'].append(random.randint(0,10))
    state["counter"] += 1
    return state

def shouldContinue(state: AgentState) -> AgentState:
    """Function to decide what to do next"""
    if state['counter'] < 5:
        return "loop"
    else:
        return "exit"
```

```python
graph = StateGraph(AgentState)

graph.add_node("Greeting",greeting)
graph.add_node("Random",randomNode)

graph.set_entry_point("Greeting")
graph.add_edge("Greeting","Random")
graph.add_conditional_edges("Random",shouldContinue,{
    "loop": "Random",
    "exit":END
})

app = graph.compile()
```

```python

result = app.invoke({"name":"Apurv","data":[]})

print(result['data'])
```
