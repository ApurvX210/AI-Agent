```python
from typing import TypedDict,List
from langgraph.graph import StateGraph
```

```python
class AgentState(TypedDict):
    data : List[int]
    name : str
    result: str

def process_values(state : AgentState) -> AgentState:
    """ This function handles multiple different inputs"""
    state['result'] = f"Hi There, {state['name']} score in this exam is {sum(state['data'])}"
    return state
```

```python
graph = StateGraph(AgentState)

graph.add_node("Process Data",process_values)
graph.set_entry_point("Process Data")
graph.set_finish_point("Process Data")

app = graph.compile()
```

```python
result = app.invoke({"data":[1,2,3,4],"name":"Apurv","result":""})
print(result["result"])
```
